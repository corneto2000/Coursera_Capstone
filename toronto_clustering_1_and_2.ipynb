1. WEB SCRAPING, WRANGLING AND CLEANING OF DATA 

```
#importing required libraries
import requests
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import collections
```

```
#retreiving all row/column values as a list

tabs = []
# File handling
with open('toronto_data.html', 'r') as fp:
    html_content = fp.read()

    table_doc = BeautifulSoup(html_content, 'html.parser')
    # parsing html content
    for tr in table_doc.table.find_all('tr'):
        tabs.append(tr.find_all('td'))
tabs.pop(0)
print(tabs)
```

```
#cleaning list content and saving in a new variable

new_tab=[]
for row in tabs:
  for elem in row:
    elem=str(elem)
    elem=elem.strip()
    elem=elem.replace('\n',' ')
    elem=elem.replace('<td>','')
    elem=elem.replace('</td>','')
    elem=elem.strip()
    new_tab.append(elem)
new_tab
```

```
#converting list data into dictionary 

my_dict=collections.defaultdict(list)
i=0
while(i<len(new_tab)):
  my_dict["Postal Code"].append(new_tab[i])
  my_dict["Borough"].append(new_tab[i+1])
  my_dict["Neighbourhood"].append(new_tab[i+2])
  i=i+3

my_dict
```

```
#converting dictionary into a dataframe

toronto_table=pd.DataFrame(my_dict)
toronto_table.head()
```

```
#deleting invalid neighbourhoods and checking shape
#Note: I dont need to join rows with similar postal codes, since they are already in that form(they weren't seperated while getting the tags.)

toronto_table=toronto_table[toronto_table.Borough!='Not assigned']
toronto_table.head()
```
